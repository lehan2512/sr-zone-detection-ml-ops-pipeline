```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import MinMaxScaler

# --- 1. Define the Hopkins Statistic Function ---
def calculate_hopkins(X):
    """
    Calculates the Hopkins Statistic for cluster tendency.
    Returns a value between 0 and 1. Values > 0.75 indicate strong clusterability.
    """
    X = np.array(X)
    d = X.shape[1]
    n = len(X)
    # Sample 10% of the 16,000 records to make computation lightning fast
    m = int(0.1 * n) 
    
    # 1. Sample m real data points
    real_indices = np.random.choice(n, m, replace=False)
    real_sample = X[real_indices]
    
    # 2. Generate m artificial uniform data points across the feature space
    mins, maxs = X.min(axis=0), X.max(axis=0)
    artificial_sample = np.random.uniform(mins, maxs, (m, d))
    
    # 3. Find distances from real points to their nearest real neighbor
    nn_real = NearestNeighbors(n_neighbors=2).fit(X)
    w_distances, _ = nn_real.kneighbors(real_sample)
    # Use the 2nd neighbor (index 1) because the 1st is the point itself (distance 0)
    w = w_distances[:, 1] ** d
    
    # 4. Find distances from artificial points to the nearest real point
    nn_artificial = NearestNeighbors(n_neighbors=1).fit(X)
    u_distances, _ = nn_artificial.kneighbors(artificial_sample)
    u = u_distances[:, 0] ** d
    
    # 5. Calculate ratio
    hopkins_score = sum(u) / (sum(u) + sum(w))
    return hopkins_score

# --- 2. Execute the Tests on your 16k Dataset ---
# Load your post-RVT filtered dataset
df = pd.read_csv('../datasets/BTCUSDT_processed.csv') # Ensure this is your 16k dataset

# Test 1: 1D Price Clustering
price_data = df[['close']].values
h_price = calculate_hopkins(price_data)

# Test 2: 2D Scaled Price + Volatility Clustering
scaler = MinMaxScaler()
price_atr_data = scaler.fit_transform(df[['close', 'atr']])
h_price_atr = calculate_hopkins(price_atr_data)

# Test 3: Synthetic Random Noise (For academic baseline)
np.random.seed(42)
random_data = np.random.uniform(0, 100, (16000, 1))
h_random = calculate_hopkins(random_data)

# --- 3. Print Results for the Report ---
print(f"--- Hopkins Statistic Results (Cluster Tendency) ---")
print(f"1D Close Price Score:           {h_price:.4f}")
print(f"2D Scaled [Close, ATR] Score:   {h_price_atr:.4f}")
print(f"Synthetic Random Baseline:      {h_random:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.preprocessing import StandardScaler

# 1. Load your sanitized, filtered dataset
df = pd.read_csv('../datasets/BTCUSDT_processed.csv')
# For S/R zones, we cluster based on Price. 
# We scale it so the math operates efficiently.
X = df[['close']].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 2. Initialize the Competitors ---
# Note: For a fair test, we use k=6 (assuming 6 major S/R zones) for KMeans/GMM.
# DBSCAN dictates its own number of clusters based on eps.

kmeans = KMeans(n_clusters=6, random_state=42, n_init=10)
gmm = GaussianMixture(n_components=6, random_state=42)
# eps=0.1 and min_samples=50 tuned roughly for scaled 1D data
dbscan = DBSCAN(eps=0.1, min_samples=50) 

# --- 3. "Train" the Models ---
labels_kmeans = kmeans.fit_predict(X_scaled)
labels_gmm = gmm.fit_predict(X_scaled)
labels_dbscan = dbscan.fit_predict(X_scaled)

# Helper function to calculate stats (ignoring noise label -1 for DBSCAN)
def get_metrics(X, labels):
    # If DBSCAN put everything in one cluster or everything is noise, return NaNs
    if len(set(labels)) <= 1 or (len(set(labels)) == 2 and -1 in labels):
        return np.nan, np.nan, np.nan
    
    # Filter out DBSCAN noise (-1) for fair metric calculation
    mask = labels != -1
    X_clean = X[mask]
    labels_clean = labels[mask]
    
    sil = silhouette_score(X_clean, labels_clean)
    db = davies_bouldin_score(X_clean, labels_clean)
    ch = calinski_harabasz_score(X_clean, labels_clean)
    return sil, db, ch

# --- 4. Evaluate ---
metrics_kmeans = get_metrics(X_scaled, labels_kmeans)
metrics_gmm = get_metrics(X_scaled, labels_gmm)
metrics_dbscan = get_metrics(X_scaled, labels_dbscan)

# --- 5. Build the Scoreboard ---
results = pd.DataFrame({
    'Model': ['K-Means', 'Gaussian Mixture', 'DBSCAN'],
    'Silhouette Score ↑': [metrics_kmeans[0], metrics_gmm[0], metrics_dbscan[0]],
    'Davies-Bouldin ↓': [metrics_kmeans[1], metrics_gmm[1], metrics_dbscan[1]],
    'Calinski-Harabasz ↑': [metrics_kmeans[2], metrics_gmm[2], metrics_dbscan[2]]
})

print("--- UNSUPERVISED MODEL EVALUATION SCOREBOARD ---")
display(results.round(4))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler

# 1. Load and prepare the data
df = pd.read_csv('../datasets/BTCUSDT_processed.csv')


X = df[['close']].values
X_scaled = StandardScaler().fit_transform(X)

# 2. Trackers for K-Means and GMM
K_range = range(2, 13) # X-Axis values
kmeans_inertia, kmeans_sil = [], []
gmm_bic, gmm_sil = [], []

for k in K_range:
    # K-Means
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans_labels = kmeans.fit_predict(X_scaled)
    kmeans_inertia.append(kmeans.inertia_)
    kmeans_sil.append(silhouette_score(X_scaled, kmeans_labels))
    
    # GMM
    gmm = GaussianMixture(n_components=k, random_state=42)
    gmm_labels = gmm.fit_predict(X_scaled)
    gmm_bic.append(gmm.bic(X_scaled)) 
    gmm_sil.append(silhouette_score(X_scaled, gmm_labels))

# 3. Plotting Academically Correct Graphs
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(2, 2, figsize=(16, 10))

# K-Means: True Elbow
ax[0,0].plot(K_range, kmeans_inertia, marker='o', color='#2c3e50', linewidth=2)
ax[0,0].set_title('K-Means: Elbow Method (Raw Inertia)', fontweight='bold')
ax[0,0].set_xlabel('Number of Clusters (K)')
ax[0,0].set_ylabel('Inertia (Variance)') # Let the large numbers show!
ax[0,0].set_xticks(K_range)

# K-Means: Silhouette
ax[0,1].plot(K_range, kmeans_sil, marker='s', color='#27ae60', linewidth=2)
ax[0,1].set_title('K-Means: Silhouette Score', fontweight='bold')
ax[0,1].set_xlabel('Number of Clusters (K)')
ax[0,1].set_ylabel('Silhouette Score')
ax[0,1].set_xticks(K_range)

# GMM: True Elbow (BIC)
ax[1,0].plot(K_range, gmm_bic, marker='o', color='#8e44ad', linewidth=2)
ax[1,0].set_title('GMM: Elbow Method (Raw BIC)', fontweight='bold')
ax[1,0].set_xlabel('Number of Components (K)')
ax[1,0].set_ylabel('BIC Score')
ax[1,0].set_xticks(K_range)

# GMM: Silhouette
ax[1,1].plot(K_range, gmm_sil, marker='s', color='#e67e22', linewidth=2)
ax[1,1].set_title('GMM: Silhouette Score', fontweight='bold')
ax[1,1].set_xlabel('Number of Components (K)')
ax[1,1].set_ylabel('Silhouette Score')
ax[1,1].set_xticks(K_range)

plt.tight_layout()
plt.show()

import numpy as np

# --- Automated Optimal K Detection ---

def find_elbow(x, y):
    """
    Finds the elbow by calculating the maximum geometric distance 
    from each point to the line connecting the first and last points.
    """
    # Coordinates of the first and last points
    p1 = np.array([x[0], y[0]])
    p2 = np.array([x[-1], y[-1]])
    
    distances = []
    for i in range(len(x)):
        p3 = np.array([x[i], y[i]])
        # Calculate the perpendicular distance from the point to the line
        distance = np.linalg.norm(np.cross(p2 - p1, p1 - p3)) / np.linalg.norm(p2 - p1)
        distances.append(distance)
        
    # The elbow is the point with the maximum distance
    optimal_index = np.argmax(distances)
    return x[optimal_index]

# Calculate Optimal K for K-Means
optimal_k_elbow_kmeans = find_elbow(list(K_range), kmeans_inertia)
optimal_k_sil_kmeans = K_range[np.argmax(kmeans_sil)]

# Calculate Optimal K for GMM
optimal_k_elbow_gmm = find_elbow(list(K_range), gmm_bic)
optimal_k_sil_gmm = K_range[np.argmax(gmm_sil)]

print("-" * 50)
print(" 🚀 AUTOMATED HYPERPARAMETER DETECTION 🚀")
print("-" * 50)
print(f"K-Means Optimal K (Elbow Method):      K = {optimal_k_elbow_kmeans}")
print(f"K-Means Optimal K (Silhouette Max):    K = {optimal_k_sil_kmeans}")
print("-" * 50)
print(f"GMM Optimal K (Elbow Method):          K = {optimal_k_elbow_gmm}")
print(f"GMM Optimal K (Silhouette Max):        K = {optimal_k_sil_gmm}")
print("-" * 50)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler

# 1. Load and scale the dataset
df = pd.read_csv('../datasets/BTCUSDT_processed.csv')
X = df[['close']].values
X_scaled = StandardScaler().fit_transform(X)

# 2. Define min_samples (Heuristic: For 16k rows in 1D, 50-100 is a strong baseline)
# We use min_samples to determine the 'k' for our NearestNeighbors test
min_samples = 50 

# 3. Calculate the distance from each point to its 50th nearest neighbor
neighbors = NearestNeighbors(n_neighbors=min_samples)
neighbors_fit = neighbors.fit(X_scaled)
distances, indices = neighbors_fit.kneighbors(X_scaled)

# Sort distances (look only at the distance to the furthest neighbor in the cluster)
distances = np.sort(distances[:, min_samples - 1], axis=0)

# --- 4. Automated Knee Detection (Finding the exact point of maximum curvature) ---
def find_knee_point(y):
    x = np.arange(len(y))
    # Line connecting first and last point
    p1 = np.array([x[0], y[0]])
    p2 = np.array([x[-1], y[-1]])
    
    distances_to_line = []
    for i in range(len(x)):
        p3 = np.array([x[i], y[i]])
        # Cross product to find perpendicular distance
        dist = np.linalg.norm(np.cross(p2 - p1, p1 - p3)) / np.linalg.norm(p2 - p1)
        distances_to_line.append(dist)
        
    return np.argmax(distances_to_line), y[np.argmax(distances_to_line)]

knee_idx, optimal_eps = find_knee_point(distances)

# --- 5. Plotting the K-Distance Graph ---
plt.figure(figsize=(12, 7))
plt.style.use('seaborn-v0_8-whitegrid')

# Plot the sorted distances
plt.plot(distances, color='#2980b9', linewidth=2.5, label='K-Distance Curve')

# Add a marker for the mathematically detected Knee
plt.scatter(knee_idx, optimal_eps, color='#e74c3c', s=100, zorder=5, 
            label=f'Optimal Epsilon (Detected Knee): {optimal_eps:.4f}')

# Add reference lines
plt.axhline(y=optimal_eps, color='#e74c3c', linestyle='--', alpha=0.7)
plt.axvline(x=knee_idx, color='#e74c3c', linestyle='--', alpha=0.7)

plt.title(f'DBSCAN Parameter Tuning: K-Distance Graph (k={min_samples})', fontsize=16, fontweight='bold')
plt.xlabel('Data Points Sorted by Distance', fontsize=12)
plt.ylabel('Epsilon Distance ($\epsilon$)', fontsize=12)
plt.legend(loc='upper left', fontsize=12)

# Zoom in on the Y-axis to make the curve visually obvious (filtering out the extreme top 1% noise)
plt.ylim(0, np.percentile(distances, 99)) 

plt.tight_layout()
plt.show()

print("-" * 50)
print(" 🚀 DBSCAN HYPERPARAMETER DETECTION 🚀")
print("-" * 50)
print(f"Targeting min_samples:  {min_samples}")
print(f"Calculated Optimal EPS: {optimal_eps:.4f}")
print("-" * 50)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler

# 1. Load and scale the dataset
df = pd.read_csv('../datasets/BTCUSDT_processed.csv')
X = df[['close']].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 2. Fit the Models with Optimized Hyperparameters
# K-Means
kmeans = KMeans(n_clusters=7, random_state=42, n_init=10)
kmeans.fit(X_scaled)
kmeans_levels = scaler.inverse_transform(kmeans.cluster_centers_).flatten()

# GMM
gmm = GaussianMixture(n_components=7, random_state=42)
gmm.fit(X_scaled)
gmm_levels = scaler.inverse_transform(gmm.means_).flatten()

# DBSCAN
dbscan = DBSCAN(eps=0.0150, min_samples=50)
db_labels = dbscan.fit_predict(X_scaled)

# Extract DBSCAN levels (calculate the mean price of each dense cluster, ignoring -1 noise)
db_levels = []
for label in set(db_labels):
    if label != -1: 
        cluster_points = X[db_labels == label]
        db_levels.append(cluster_points.mean())
db_levels = np.array(db_levels)

# Sort all levels from lowest price to highest price
kmeans_levels.sort()
gmm_levels.sort()
db_levels.sort()

# 3. Plotting the Ultimate Visual Comparison
fig, axes = plt.subplots(3, 1, figsize=(16, 20), sharex=True)
plt.style.use('seaborn-v0_8-whitegrid')

# Helper function to plot zones cleanly
def plot_zones(ax, levels, title, line_color):
    # Plot the raw price
    ax.plot(df.index, df['close'], color='#2c3e50', alpha=0.4, label='BTC/USDT Close Price', linewidth=1.5)
    
    # Plot the structural lines
    for i, level in enumerate(levels):
        ax.axhline(y=level, color=line_color, linestyle='--', linewidth=3, alpha=0.9, 
                   label='Detected S/R Zone' if i == 0 else "") # Only label once to avoid legend clutter
        
    ax.set_title(title, fontsize=16, fontweight='bold')
    ax.set_ylabel('Price (USD)', fontsize=12)
    ax.yaxis.set_major_formatter(plt.matplotlib.ticker.StrMethodFormatter('${x:,.0f}'))
    ax.legend(loc='upper right', fontsize=12)

# Graph 1: K-Means (Distance-based)
plot_zones(axes[0], kmeans_levels, '1. K-Means Optimal Structural Zones (K=7)', '#e74c3c')

# Graph 2: GMM (Probabilistic)
plot_zones(axes[1], gmm_levels, '2. Gaussian Mixture Model Structural Zones (K=7)', '#8e44ad')

# Graph 3: DBSCAN (Density-based)
plot_zones(axes[2], db_levels, f'3. DBSCAN Density Nodes (eps=0.0150, min_samples=50)\nFound {len(db_levels)} Structural Zones', '#27ae60')

plt.xlabel('Chronological Time (Hourly Candles)', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.preprocessing import StandardScaler

# --- 1. Load and Scale Data ---
df = pd.read_csv('../datasets/BTCUSDT_processed.csv')
X = df[['close']].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# --- 2. Initialize the Optimized Models ---
kmeans = KMeans(n_clusters=7, random_state=42, n_init=10)
gmm = GaussianMixture(n_components=7, random_state=42)
dbscan = DBSCAN(eps=0.0150, min_samples=50) 

# --- 3. Fit and Predict ---
labels_kmeans = kmeans.fit_predict(X_scaled)
labels_gmm = gmm.fit_predict(X_scaled)
labels_dbscan = dbscan.fit_predict(X_scaled)

# --- 4. Final Evaluation Scoreboard ---
def get_final_metrics(X, labels):
    # Filter out DBSCAN noise (-1) to ensure fair mathematical comparison
    mask = labels != -1
    if len(set(labels[mask])) < 2:
        return np.nan, np.nan, np.nan
    X_clean = X[mask]
    labels_clean = labels[mask]
    
    sil = silhouette_score(X_clean, labels_clean)
    db = davies_bouldin_score(X_clean, labels_clean)
    ch = calinski_harabasz_score(X_clean, labels_clean)
    return sil, db, ch

metrics_kmeans = get_final_metrics(X_scaled, labels_kmeans)
metrics_gmm = get_final_metrics(X_scaled, labels_gmm)
metrics_dbscan = get_final_metrics(X_scaled, labels_dbscan)

results = pd.DataFrame({
    'Model': ['K-Means (K=7)', 'GMM (K=7)', 'DBSCAN (eps=0.0150, min=50)'],
    'Silhouette Score ↑': [metrics_kmeans[0], metrics_gmm[0], metrics_dbscan[0]],
    'Davies-Bouldin ↓': [metrics_kmeans[1], metrics_gmm[1], metrics_dbscan[1]],
    'Calinski-Harabasz ↑': [metrics_kmeans[2], metrics_gmm[2], metrics_dbscan[2]]
})

print("-" * 65)
print(" 🏆 FINAL UNSUPERVISED MODEL SCOREBOARD 🏆")
print("-" * 65)
display(results.round(4))
print("-" * 65)

# --- 5. Visualizing the Winning Architecture (K-Means) ---
# Extract centroids and convert back to actual BTC prices
usd_centroids = scaler.inverse_transform(kmeans.cluster_centers_).flatten()
usd_centroids.sort() 

plt.figure(figsize=(16, 8))
plt.style.use('seaborn-v0_8-whitegrid')

# Plot the raw BTC Close Price
plt.plot(df.index, df['close'], color='#34495e', alpha=0.8, linewidth=1, label='BTC/USDT Close Price')

# Overlay the mathematically extracted S/R Zones
colors = ['#e74c3c', '#e67e22', '#f1c40f', '#2ecc71', '#1abc9c', '#9b59b6', '#ea2027']
for i, price in enumerate(usd_centroids):
    plt.axhline(y=price, color=colors[i], linestyle='-', linewidth=3, alpha=0.8, 
                label=f'Structural Zone {i+1}: ${price:,.0f}')

plt.title('Algorithmic Market Structure: Optimal Support & Resistance Zones', fontsize=18, fontweight='bold')
plt.xlabel('Time (Hourly Candles)', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)

# Format y-axis to show dollar values clearly
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.StrMethodFormatter('${x:,.0f}'))
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), title="Extracted S/R Levels")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load both the raw and the processed datasets
# (Ensure these paths match your directory structure)
df_raw = pd.read_csv('../datasets/BTCUSDT_raw.csv')
df_filtered = pd.read_csv('../datasets/BTCUSDT_processed.csv')

# 2. Calculate the specific Standard Deviation metrics
raw_std = df_raw['volume'].std()
filtered_std = df_filtered['volume'].std()
reduction_pct = ((raw_std - filtered_std) / raw_std) * 100

print(f"Raw Volume Std Dev:      {raw_std:,.2f}")
print(f"Filtered Volume Std Dev: {filtered_std:,.2f}")
print(f"Total Reduction:         {reduction_pct:.1f}%")

# 3. Create the Before-and-After Visualizations
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
plt.style.use('seaborn-v0_8-whitegrid')

# --- Graph 1: Boxplot Comparison (Outlier Reduction) ---
# Combine data for seaborn plotting
raw_data = pd.DataFrame({'Volume': df_raw['volume'], 'State': '1. Raw Data'})
filtered_data = pd.DataFrame({'Volume': df_filtered['volume'], 'State': '2. Post-RVT Filter'})
combined_df = pd.concat([raw_data, filtered_data])

sns.boxplot(x='State', y='Volume', data=combined_df, ax=axes[0], palette=['#e74c3c', '#2ecc71'])
axes[0].set_title('Volume Outlier Reduction (Boxplot)', fontweight='bold', fontsize=14)
axes[0].set_ylabel('Trading Volume (BTC)')
axes[0].set_yscale('log') # Log scale to make the massive raw outliers visible

# --- Graph 2: KDE Density Comparison (Distribution Stabilization) ---
sns.kdeplot(df_raw['volume'], ax=axes[1], color='#e74c3c', fill=True, label=f'Raw ($\sigma$ = {raw_std:,.0f})', alpha=0.3)
sns.kdeplot(df_filtered['volume'], ax=axes[1], color='#2ecc71', fill=True, label=f'Filtered ($\sigma$ = {filtered_std:,.0f})', alpha=0.5)

axes[1].set_title('Volume Distribution Stability (KDE)', fontweight='bold', fontsize=14)
axes[1].set_xlabel('Trading Volume (BTC)')
axes[1].set_ylabel('Density')
axes[1].set_xlim(0, df_filtered['volume'].quantile(0.99) * 2) # Zoom in to ignore the extreme 1% raw tail for readability
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the raw dataset
df_raw = pd.read_csv('../datasets/BTCUSDT_raw.csv')

# Drop any non-numeric columns if they exist (like a timestamp string)
# df_raw = df_raw.select_dtypes(include=[np.number])

# 2. Calculate the Pearson Correlation Matrix
corr_matrix = df_raw.corr()

# 3. Create a mask to hide the upper triangle (makes the chart cleaner)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# 4. Set up the matplotlib figure
plt.figure(figsize=(12, 10))
plt.style.use('seaborn-v0_8-whitegrid')

# 5. Draw the heatmap
sns.heatmap(
    corr_matrix, 
    mask=mask, 
    cmap='coolwarm',       # A great diverging color map (Blue = Negative, Red = Positive)
    vmax=1.0, vmin=-1.0,   # Correlation is strictly between -1 and 1
    center=0,              # Center the color map at 0
    annot=True,            # Show the actual numbers inside the boxes
    fmt=".2f",             # Format numbers to 2 decimal places
    square=True, 
    linewidths=.5, 
    cbar_kws={"shrink": .75} # Make the colorbar slightly smaller
)

plt.title('Feature Correlation Heatmap (Raw Dataset)', fontsize=16, fontweight='bold', pad=20)
plt.xticks(rotation=45, ha='right', fontsize=11)
plt.yticks(rotation=0, fontsize=11)

plt.tight_layout()
plt.show()
```
